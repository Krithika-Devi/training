**Write a program, where you will create a parent class (eg. called Person), and you will at least 3 different child classes, and use any kind of loop (eg. to estimate atleast 2  attributes mean, strengths, and weakness of the person.)**

### Hybrid Inheritance


```python
# parent class

class Product(object):
    
    # class variable
    P1 = 'Amazon'
    P2 = 'Flipkart'
    
    def __init__(self,productid,productname):
        # instance variable
        self.productid = productid
        self.productname = productname
    
    def OriginalPrice(self,price):
        self.price = price
        print('Product id: ', self.productid)
        print('Product Name: ', self.productname)
        print('Product Price: ', self.price)
        
    def DiscountPercentage(self,discountpercentage):
        self.discountpercentage = discountpercentage
        print('Discount Percentage: ', self.discountpercentage)
        
    def SellingPrice(self):
        self.sellingprice = self.price - (self.price*self.discountpercentage/100)
        print('Selling Price: ', self.sellingprice)
```

```python
# child class 1 --> created on parent class

class vendor_central(Product): 
    
    def __init__(self,productid,productname,country):
        self.country = country
        Product.__init__(self,productid,productname)
        
    def WeekSales(self):
        self.days =7
        self.sales = [] 
        self.total = 0
        for i in range(self.days):
            self.sales.append(int(input('Enter total sales per day: ')))
            self.total += self.sales[i]
        print("\n Total sales of the week - ",self.total)     
```

```python
# child class 2 --> created on parent class

class seller_central(Product):
    
    def __init__(self,productid,productname,reviewstar):
        self.reviewstar = reviewstar
        print("Number of stars received: ", self.reviewstar)
        Product.__init__(self,productid,productname)
    
    def reviews(self):
        if self.reviewstar <=2:
            print('Bad Product')
        elif self.reviewstar == 3:
            print('Average Product')
        else:
            return('Good Product')
```

```python
# child class 3 --> created on child class 1

class customer(vendor_central):
    
    def __init__(self,productid,productname,country,name):
        self.name = name
        vendor_central.__init__(self,productid,productname,country)
    
    def welcome(self):
        print('Hi, ' + self.name + ', Welcome!')
```

```python
prod = Product(32,'watch') # object on parent class

print(prod.P1)
prod.OriginalPrice(6200)
prod.DiscountPercentage(15)
prod.SellingPrice()
```

```python
vendor = vendor_central('32KD80','Allen Solly','India') # object on child class 1

print(vendor.P2,'\n')
vendor.WeekSales()
```

```python
seller = seller_central('33FKC8','ShoeBuy',3) # object on child class 2

seller.reviews()
```

```python
cus = customer('44VC15','Audible Books','India','Krithika') # object on child class 3

cus.welcome()
```
